```python
import pandas as pd
```

```python
df = pd.read_csv('data.csv')
```

```python
df.head()
```

```python
df = df.drop(columns=['Unnamed: 32','id'])
```

```python
df.info()
```

```python
df.describe()
```

```python
import numpy as np
```

```python
df['diagnosis'].value_counts()
```

```python
df = df.replace(['M','B'],[0,1])
```

```python
from scipy import stats


df = df[(np.abs(stats.zscore(df)) < 3).all(axis=1)]
```

```python
df['diagnosis'].value_counts()
```

```python
df.drop_duplicates()
```

```python
X = df.drop(columns='diagnosis')
```

```python
from sklearn.cluster import KMeans, DBSCAN
```

```python
kmeans = KMeans(n_clusters=2,n_init='auto')
```

```python
kmeans.fit(X)
```

```python
from sklearn.metrics import adjusted_rand_score,silhouette_score
```

```python
ARI_kmeans = adjusted_rand_score(df['diagnosis'],kmeans.labels_)
```

```python
ARI_kmeans
```

```python
silhouette_score(X,kmeans.labels_)
```

```python
db = DBSCAN()
```

```python
db.fit(X)
```

```python
set(db.labels_)
```

```python
ari_max = 0
best_ep = 0
best_samp = 0

for k in range(10,70,2):
    for samp in range(2,9):
        db = DBSCAN(eps = k, min_samples=samp)
        db.fit(X)
        ARI = adjusted_rand_score(df['diagnosis'],db.labels_)
        if ARI > ari_max:
            ari_max = ARI
            best_ep = k
            best_samp = samp

print(ari_max)
print(best_ep)
print(best_samp)
```

```python
from sklearn.preprocessing import StandardScaler,MinMaxScaler
```

```python
X_stand = StandardScaler().fit_transform(X)
X_minmax = MinMaxScaler().fit_transform(X)
```

```python
kmeans_stand = KMeans(n_clusters=2,n_init='auto')
kmeans_minmax = KMeans(n_clusters=2,n_init='auto')
```

```python
kmeans_stand.fit(X_stand)
kmeans_minmax.fit(X_minmax)
```

```python
ARI_stand = adjusted_rand_score(df['diagnosis'],kmeans_stand.labels_)
ARI_minmax = adjusted_rand_score(df['diagnosis'],kmeans_minmax.labels_)
print(ARI_stand)
print(ARI_minmax)
```

```python
fig,(ax1,ax2,ax3,ax4)= plt.subplots(1,4,figsize=(10,10))
ax1.scatter(pca_comp[:,0],pca_comp[:,1],c=df['diagnosis'])
ax2.scatter(pca_comp[:,0],pca_comp[:,1],c=kmeans.labels_)
ax3.scatter(pca_comp[:,0],pca_comp[:,1],c=kmeans_stand.labels_)
ax4.scatter(pca_comp[:,0],pca_comp[:,1],c=kmeans_minmax.labels_)

fig.show()
```

```python
ari_max_stand = 0
best_ep_stand = 0
best_samp_stand = 0

for k in np.arange(0.3,5,0.1):
    for samp in range(2,9):
        db_stand = DBSCAN(eps = k, min_samples=samp)
        db_stand.fit(X_stand)
        ARI = adjusted_rand_score(df['diagnosis'],db_stand.labels_)
        if ARI > ari_max_stand:
            ari_max_stand = ARI
            best_ep_stand = k
            best_samp_stand = samp

print(ari_max_stand)
print(best_ep_stand)
print(best_samp_stand)
```

```python
db_stand = DBSCAN(eps=best_ep_stand,min_samples=best_samp_stand)
db_stand.fit(X_stand)
adjusted_rand_score(df['diagnosis'],db_stand.labels_)
```

```python
ari_max_mm = 0
best_ep_mm = 0
best_samp_mm = 0

for k in np.arange(0.1,2,0.1):
    for samp in range(2,9):
        db_mm = DBSCAN(eps = k, min_samples=samp)
        db_mm.fit(X_minmax)
        ARI = adjusted_rand_score(df['diagnosis'],db_mm.labels_)
        if ARI > ari_max_mm:
            ari_max_mm = ARI
            best_ep_mm = k
            best_samp_mm = samp

print(ari_max_mm)
print(best_ep_mm)
print(best_samp_mm)
```

```python
db_mm= DBSCAN(eps = best_ep_mm, min_samples=best_samp_mm)
db_mm.fit(X_minmax)
```
